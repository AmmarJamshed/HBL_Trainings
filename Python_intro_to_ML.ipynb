In this notebook, we apply two regression models and two classification models on the churn dataset to evaluate their performance.

## Regression Models

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
df = pd.read_csv('/content/Churn_Modelling.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# Data Preprocessing
df['Geography'] = LabelEncoder().fit_transform(df['Geography'])
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])

# Splitting the dataset
X_reg = df[['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember']]
y_reg = df['EstimatedSalary']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
# Regression Model 1: Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train_reg, y_train_reg)
y_pred_linear = linear_model.predict(X_test_reg)
```

```python
# Regression Model 2: Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_reg, y_train_reg)
y_pred_rf = rf_model.predict(X_test_reg)
```

```python
# Evaluation
## Model 1
linear_mse = mean_squared_error(y_test_reg, y_pred_linear)
linear_r2 = r2_score(y_test_reg, y_pred_linear)
## Model 2
rf_mse = mean_squared_error(y_test_reg, y_pred_rf)
rf_r2 = r2_score(y_test_reg, y_pred_rf)

(linear_mse, linear_r2, rf_mse, rf_r2)
```

## Classification Models

```python

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Splitting the dataset
X_clf = df[['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember']]
y_clf = df['Exited']
```

```python
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# Feature Scaling for classification
scaler = StandardScaler()
X_train_clf = scaler.fit_transform(X_train_clf)
X_test_clf = scaler.transform(X_test_clf)
```

```python
# Classification Model 1: Logistic Regression
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_clf, y_train_clf)
y_pred_logistic = logistic_model.predict(X_test_clf)

# Classification Model 2: SVM
svm_model = SVC()
svm_model.fit(X_train_clf, y_train_clf)
y_pred_svm = svm_model.predict(X_test_clf)
```

```python
# Evaluation
logistic_accuracy = accuracy_score(y_test_clf, y_pred_logistic)
svm_accuracy = accuracy_score(y_test_clf, y_pred_svm)
logistic_report = classification_report(y_test_clf, y_pred_logistic)
svm_report = classification_report(y_test_clf, y_pred_svm)

print(logistic_accuracy, svm_accuracy, logistic_report, svm_report)
```
